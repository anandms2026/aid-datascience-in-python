```python
import re
def grades():
    with open ("grades.txt", "r") as file:
    grades = file.read()
    print(grades)
    pattern = r'\b([^:\n]*):\s*B'
    return re.findall(pattern, grades)
    raise NotImplementedError()
try:
    assert len(grades()) == 16
except NotImplementedError as e:
    print(f"Error caught: {e}")
```

```python
import re
def logs():
    with open("logdata.txt", "r") as file:
        logdata = file.read()
    log_entries = []
    # print(logdata)
    # pattern = r'"host":"([^"]*)",\s*"user_name":"([^"]*)",\s*"time":"([^"]*)",\s*"request":"([^"]*)"'
    pattern = re.compile(r'^(?P<host>\S+) - (?P<user_name>\S+) \[(?P<time>[^\]]+)\] "(?P<request>[^"]+)"')
    for line in logdata.strip().split('\n'):
        match = pattern.match(line)
        # print(match)
        if match:
            log_entries.append(match.groupdict())
    return log_entries

try:
    assert len(logs()) == 979
except NotImplementedError as e:
    print(e)
```
