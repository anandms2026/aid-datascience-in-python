```python
import csv
from itertools import islice
%precision 2
def proportion_of_education():
    edu_counts = {
        1: 0, # less than high school
        2: 0, # high school
        3: 0, # more than high school but not college
        4: 0  # college
    }
    # Mapping for the final output keys
    edu_map = {
        1: "less than high school",
        2: "high school",
        3: "more than high school but not college",
        4: "college"
    }
    row_count=0
    try:
        with open('datasets/NISPUF17.csv') as csvfile:
            reader=csv.DictReader(csvfile)
            for row in reader:
                #print(f"Row {i + 1}: {row}")
                try:
                    educ_level = int(row['EDUC1'])
                    if educ_level in edu_counts:
                            edu_counts[educ_level] += 1
                except ValueError: 
                    continue

            total_children = sum(edu_counts.values())   
            proportions = {}
            for k, v in edu_counts.items():
                cat_name=edu_map[k]
                proportions[cat_name] = round(v / total_children, 2)
            print(proportions)
            return proportions
    except FileNotFoundError:
        print('Check the files needs to viewed')
    
```
