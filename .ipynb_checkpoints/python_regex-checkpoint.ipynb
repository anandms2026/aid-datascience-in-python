```python
################################# findall()#########################
## re.findall(pattern, string, flags=0) #####################
##Return all non-overlapping matches of pattern in List/Tuple, as a list of strings or tuples.
text = "The quick brown fox jumps over the lazy dog. The dog is brown.There are 123 apples and 45 oranges."
# Find all occurrences of "the" (case-sensitive)
re.findall(r"the", text)
# Find all occurrences of "the" (case-insensitive)
matches_case_insensitive = re.findall(r"the", text, re.IGNORECASE)
print(matches_case_insensitive)
# Find all words starting with 'b'  
#\w -- is used for words pattern
#\d -- is used for digits pattern
#\s -- is used for space pattern
#() -- is used for capturing groups
#* -- is used for zero or more occurances of the preceeding character
#. -- is used for any Match any character
#^ -- is used Match the start of the string
#+ -- is used for one or more of the preceding element

#re.findall(r"[a-zA-Z]{1,100}\\FERPA\\",wiki)


print(re.findall(r"edit\s",wiki)) #Pattern followed by space
print(re.findall(r"edit\s+\w ",wiki)) #Pattern followed by space and next one word
print(re.findall(r"edit\s+(\w+)",wiki)) #the word after the pattern search

#search_pattern_python="\b\w+\b{1,100}\brFERPA\b"
#re.findall(search_pattern_python,wiki)
#print("length")
#len(re.findall("[a-zA-Z]{1,100}",wiki))
#text = "A cat sat on the mat."

#pattern = r"cat[\w\d]*(?=\s)"
#re.findall(pattern, text)

#text = "Hello World! How are you?"
#pattern = r"Hello\s+(\w+)"

#re.findall(pattern, text)

color = "Color: red, Size: large, Type: fruit"
# Find values associated with 'Color', 'Size', or 'Type'
pattern = r"Color: (.*?)|Size: (.*?)|Type: (.*?)"
matches = re.findall(pattern, color)
print(f"values associated with 'Color', 'Size', or 'Type':::::::::",matches)

text2 = "She said, \"Hello there!\" and then added, \"How are you?\""
pattern = r'"(.*?)"'  # Non-greedy match of any character inside quotes
matches = re.findall(pattern, text2)
print(f"text2:::::::::::",matches)


words_starting_with_b = re.findall(r"\bl\w+", text, re.IGNORECASE)
print(f"Words starting with 'l': {words_starting_with_b}")

# Find all numbers
numbers = re.findall(r"\d+", text)
print(f"Numbers: {numbers}")

# Find matches with capturing groups
sentence = "Name: Alice, Age: 30; Name: Bob, Age: 25"
name_age_pairs = re.findall(r"Name: (\w+), Age: (\d+)", sentence)
print(f"Name and age pairs: {name_age_pairs}")
```
